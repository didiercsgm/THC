```python
def convertlen(): #definición de la función para convertir unidades de longitud

  print("Introduce las unidades a convertir") #texto de instrucción al usuario
  print("Las opciones son: m, cm, mm, km, milla, yarda, pie, nm (milla náutica) y furlong")
  print("")

  units = {"m":1.0,"cm":0.01,"mm":0.001,"km":1000.0,"milla":1609.344,"yarda":0.9144,"pie":0.3048,"nm":1852.0,"furlong":201.168} #diccionario que contiene a todas las unidades que se pueden convertir como múltiplos de una unidad SI base

  while True: #le pide al usuario las unidades que quiere convertir. el bucle existe para que en el caso de que el input no sea válido, lo pida de nuevo hasta que el input sea miembro del diccionario
    first = input("de qué unidad quieres convertir? ")
    if first in units: #revisa si el input está en el diccionario
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue #empieza desde nuevo el bucle

  while True: #hace lo mismo que el bucle anterior pero para la unidad a la cual se va a convertir
    to = input("a qué unidad quieres convertir? ")
    if to in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  print("introduce la cantidad a convertir")

  check = False
  while not check: #este bucle sirve para asegurarse que la cantidad de unidades sea un número
    try:
      qty = float(input())
      check = True
    except ValueError: #en la línea 32 intenta convertir el str del input a un float, esto solo se puede si son números, si no son números, da un error; el except sirve para que cuando aparezca el error solicita el número de nuevo hasta que sea número
         print("no es un número, intenta de nuevo")

  res1 = float((units[first]/units[to])) # divide la cantidad a convertirse a la unidad convertida, esto lo hace dividiendo la primera por la segunda. sirve para cualquier unidad con cualquier tamaño con que esté el el diccionario
  res2 = qty*res1
  print(qty,first,"son", round(res2, 5),to) #imprime el resultado convertido


def convertmass(): # sirve igual que el anterior, pero con un diccionario cambiado para la masa

  print("Introduce las unidades a convertir")
  print("Las opciones son: kg, g, mg (tonelada métrica), oz (onza), lb (libra), st (stone), lt (tonelada larga) y tn (tonelada corta)")
  print("")

  units = {"kg":1.0,"g":0.001,"mg":1000.0,"oz":0.0283495,"lb":0.45359237,"st":6.35029318,"lt":1016.047,"tn":907.185}

  while True:
    first = input("de qué unidad quieres convertir? ")
    if first in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  while True:
    to = input("a qué unidad quieres convertir? ")
    if to in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  print("introduce la cantidad a convertir")

  check = False
  while not check:
    try:
      qty = float(input())
      check = True
    except ValueError:
         print("no es un número, intenta de nuevo")

  res1 = float((units[first]/units[to]))
  res2 = qty*res1
  print(qty,first,"son", round(res2, 5),to)


def converttime(): # sirve igual que el anterior, pero con un diccionario cambiado para el tiempo
  print("Introduce las unidades a convertir")
  print("Las opciones son: s (segundo), ms (milisegundo), m (minuto), h (hora), d (día), se (semana), ct (cuarentena), a (año) y ab (año bisiesto)")
  print("")

  units = {"s":1.0,"ms":0.001,"m":60.0,"h":3600.0,"d":86400.0,"se":6.35029318,"ct":3456000.0,"a":31536000,"ab":31622400}

  while True:
    first = input("de qué unidad quieres convertir? ")
    if first in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  while True:
    to = input("a qué unidad quieres convertir? ")
    if to in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  print("introduce la cantidad a convertir")

  check = False
  while not check:
    try:
      qty = float(input())
      check = True
    except ValueError:
         print("no es un número, intenta de nuevo")

  res1 = float((units[first]/units[to]))
  res2 = qty*res1
  print(qty,first,"son", round(res2, 5),to)


def convertvol(): # sirve igual que el anterior, pero con un diccionario cambiado para el volumen
  print("Introduce las unidades a convertir")
  print("Las opciones son: l (litro), ml (mililitro), mc (metro cúbico), gl (galón), pt (pinta), ol (onza líquida) y bl (barríl)")
  print("")

  units = {"l":1.0,"ml":0.001,"mc":1000.0,"gl":3.785411784,"pt":0.473176473,"ol":6.35029318,"bl":158.987294928}

  while True:
    first = input("de qué unidad quieres convertir? ")
    if first in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  while True:
    to = input("a qué unidad quieres convertir? ")
    if to in units:
      True
      break
    else:
      print("debes introducir una de las opciones dadas, intenta de nuevo")
    continue

  print("introduce la cantidad a convertir")

  check = False
  while not check:
    try:
      qty = float(input())
      check = True
    except ValueError:
         print("no es un número, intenta de nuevo")

  res1 = float((units[first]/units[to]))
  res2 = qty*res1
  print(qty,first,"son", round(res2, 5),to)


def converter(): # esta función adjunta a todas las anteriores funciones; además crea el menu principal

  print("Bienvenido al convertidor de unidades, deseas convertir unidades de distancia, masa o tiempo?")
  print("Si quieres salir del programa, pon: salir")
  print("")

  while True: # este bucle existe para tomar el tipo de unidad que se va a convertir y para asegurarse que el input sea válido
    what = input()
    if what == "distancia":
      convertlen() #llama la función porque el usuario la pidió
      True
      break
    elif what == "masa":
      convertmass()
      True
      break
    elif what == "tiempo":
      converttime()
      True
      break
    elif what == "volumen":
      convertvol()
      True
      break
    elif what == "salir": #está para poder salirse del convertidor
      print("adiós") #se despide
      return
    else:
      print("no es válido, intenta de nuevo")
      continue

  print("deseas convertir algo más? si o no")

  while True: #el bucle da la opción de convertir otra unidad, o si ya salir del programa
    more = input()
    if more == "si":
      converter()
      True
      break
    elif more == "no":
      print("adiós")
      return
    else:
      print("debes decir si o no")
    continue


converter()
```
