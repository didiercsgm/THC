```python
import matplotlib.pyplot as plt
import numpy as np

angbase = np.linspace(1, 2*np.pi, 101)
angs  = angbase*np.pi

sin = np.sin(angs)
plt.subplot(2, 3, 1)
plt.plot(angs,sin)

cos = np.cos(angs)
plt.subplot(2, 3, 2)
plt.plot(angs,cos)

tan = np.tan(angs)
plt.subplot(2, 3, 3)
plt.plot(angs,tan)

csc = 1/sin
plt.subplot(2, 3, 4)
plt.plot(angs,csc)

sec = 1/cos
plt.subplot(2, 3, 5)
plt.plot(angs,sec)

cot = 1/tan
plt.subplot(2, 3, 6)
plt.plot(angs,cot)

plt.show()
```

```python
x = np.linspace(1,100,100)

y = np.zeros(100)

running_sum = 0

last_ds = np.zeros(100)

for i in range(len(y)):
  newnum= np.random.randint(-10,10)
  running_sum += newnum
  y[i] = running_sum
  last_ds[i] = y[i] % 10

plt.plot(x,y)
plt.subplot(1,1,1)
plt.show
```
